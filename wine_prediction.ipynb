```python
red['type'] = 1

# Add `type` column to `white` with value 0
white['type'] = 0

# Append `white` to `red`
wines = red.append(white, ignore_index=True)
```

```python
print(wines)
```

```python
import seaborn as sns
corr = wines.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python
# Import `train_test_split` from `sklearn.model_selection`
import numpy as np
from sklearn.model_selection import train_test_split

# Specify the data 
X=wines.ix[:,0:11]

# Specify the target labels and flatten the array 
y=np.ravel(wines.type)

# Split the data up in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Define the scaler 
scaler = StandardScaler().fit(X_train)

# Scale the train set
X_train = scaler.transform(X_train)

# Scale the test set
X_test = scaler.transform(X_test)
```

```python
print(X_train)
print("..................................................................")
print(X_test)
```

```python
# Import `Sequential` from `keras.models`
from keras.models import Sequential

# Import `Dense` from `keras.layers`
from keras.layers import Dense

# Initialize the constructor
model = Sequential()

# Add an input layer 
model.add(Dense(12, activation='relu', input_shape=(11,)))

model.add(Dense(12, activation='relu'))

model.add(Dense(12, activation='relu'))
# Add one hidden layer 
model.add(Dense(8, activation='relu'))
#  2
model.add(Dense(8, activation='relu'))
#  3
model.add(Dense(8, activation='relu'))

# Add an output layer 
model.add(Dense(1, activation='sigmoid'))
```

```python
# Model output shape
model.output_shape
```

```python

# Model summary
model.summary()
#output_size * (input_size + 1) == number_parameters
```

```python
# Model config
model.get_config()
```

```python

# List all weight tensors 
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
                   
```

```python
model.fit(X_train, y_train,epochs=30, batch_size=10, verbose=1)
```

```python
y_pred = model.predict(X_test)
print(y_pred[1])
#y_pred=np.asarray(y_pred)# here we convert an y_pred list into an numpy array
#print("out put belonge to this class:",np.argmax(y_pred,axis=0))#row wise check karega maxumum no ka index return karega column wise.
```

```python
score = model.evaluate(X_test, y_test,verbose=1)

print(score)
```
